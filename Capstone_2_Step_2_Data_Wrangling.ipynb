## Import 2 years of Loopring data from CryptoCompare.com

Import 2 years of hourly Loopring (LRC) data.

There is a limit of 2000 hour calls per request so will have to make 5 separate requests.

```python
# API key for CryptoCompare.com

load_dotenv()
API_KEY = os.getenv('CRYPTOCOMPARE_API')
```

```python
# API connection
url = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym=LRC&tsym=USD&limit=2000&api_key=API_KEY'
r = requests.get(url)
json_data = r.json()

# Save dataset for LRC in json format
with open("LRC_file.json", "w") as write_file:
    json.dump(json_data, write_file)
```

```python
# 2nd batch of data

url = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym=LRC&tsym=USD&limit=2000&toTs=1634846400&api_key=API_KEY'
r = requests.get(url)
json_data = r.json()

# Save dataset for LRC in json format
with open("LRC_file2.json", "w") as write_file:
    json.dump(json_data, write_file)
```

```python
# 3rd batch of data

url = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym=LRC&tsym=USD&limit=2000&toTs=1627646400&api_key=API_KEY'
r = requests.get(url)
json_data = r.json()

# Save dataset for LRC in json format
with open("LRC_file3.json", "w") as write_file:
    json.dump(json_data, write_file)
```

```python
# 4th batch of data

url = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym=LRC&tsym=USD&limit=2000&toTs=1620446400&api_key=API_KEY'
r = requests.get(url)
json_data = r.json()

# Save dataset for LRC in json format
with open("LRC_file4.json", "w") as write_file:
    json.dump(json_data, write_file)
```

```python
# 5th batch of data

url = 'https://min-api.cryptocompare.com/data/v2/histohour?fsym=LRC&tsym=USD&limit=1000&toTs=1613246400&api_key=API_KEY'
r = requests.get(url)
json_data = r.json()

# Save dataset for LRC in json format
with open("LRC_file5.json", "w") as write_file:
    json.dump(json_data, write_file)
```

```python
# convert file data into dictionaries

with open("LRC_file.json", "r") as lrc1:
    lrc1 = json.load(lrc1)
    
with open("LRC_file2.json", "r") as lrc2:
    lrc2 = json.load(lrc2)

with open("LRC_file3.json", "r") as lrc3:
    lrc3 = json.load(lrc3)

with open("LRC_file4.json", "r") as lrc4:
    lrc4 = json.load(lrc4)

with open("LRC_file5.json", "r") as lrc5:
    lrc5 = json.load(lrc5)
```

```python
# Strip down to just the hour, high & low
lrc1_data = lrc1['Data']['Data'][:]
lrc2_data = lrc2['Data']['Data'][:]
lrc3_data = lrc3['Data']['Data'][:]
lrc4_data = lrc4['Data']['Data'][:]
lrc5_data = lrc5['Data']['Data'][:]
```

```python
# Combine data from the 5 batches and structure

lrc_data = {}
lrc_time = []
lrc_open = []
lrc_close = []


for i in lrc5_data:
    lrc_hr = datetime.datetime.fromtimestamp(int(i['time'])).strftime('%Y-%m-%d %H:%M')
    lrc_time.append(lrc_hr)
    lrc_open.append(i['open'])
    lrc_close.append(i['close'])
    lrc_data={'date':lrc_time, 'open':lrc_open, 'close':lrc_close}
 

for i in lrc4_data:
    lrc_hr = datetime.datetime.fromtimestamp(int(i['time'])).strftime('%Y-%m-%d %H:%M')
    lrc_time.append(lrc_hr)
    lrc_open.append(i['open'])
    lrc_close.append(i['close'])
    lrc_data={'date':lrc_time, 'open':lrc_open, 'close':lrc_close}
 

for i in lrc3_data:
    lrc_hr = datetime.datetime.fromtimestamp(int(i['time'])).strftime('%Y-%m-%d %H:%M')
    lrc_time.append(lrc_hr)
    lrc_open.append(i['open'])
    lrc_close.append(i['close'])
    lrc_data={'date':lrc_time, 'open':lrc_open, 'close':lrc_close}
 

for i in lrc2_data:
    lrc_hr = datetime.datetime.fromtimestamp(int(i['time'])).strftime('%Y-%m-%d %H:%M')
    lrc_time.append(lrc_hr)
    lrc_open.append(i['open'])
    lrc_close.append(i['close'])
    lrc_data={'date':lrc_time, 'open':lrc_open, 'close':lrc_close}
    

for i in lrc1_data:
    lrc_hr = datetime.datetime.fromtimestamp(int(i['time'])).strftime('%Y-%m-%d %H:%M')
    lrc_time.append(lrc_hr)
    lrc_open.append(i['open'])
    lrc_close.append(i['close'])
    lrc_data={'date':lrc_time, 'open':lrc_open, 'close':lrc_close}
```

```python
# Convert lrc_data into a DataFrame

lrc_df = pd.DataFrame(lrc_data)
print(lrc_df.head())
```

## Scrape Reddit data

Pushshift.io's API will be used to scrape the cryptocurrency and superstonk subreddits for comment data.

pmaw api wrapper isn't available for installation on local machine using anaconda so a notebook on google colab will be used to scrape data and then saved to csv format for further processing on local machine.

```python
# instance of PushshiftAPI w/ default parameters
api = PushshiftAPI()

# start stop timestamps
before = int(datetime.datetime(2022,1,1,0,0).timestamp())
after = int(datetime.datetime(2020,1,1,0,0).timestamp())
```

```python
# scrape superstonk subreddit for comments mentioning 'loopring'

subreddit='superstonk'
limit=300000
filters=['created_utc', 'body', 'id']
loopring_ss_comments = api.search_comments(q='loopring', subreddit=subreddit,filter=filters, limit=limit, before=before, after=after)
print(f'Retrieved {len(loopring_ss_comments)} comments from Pushshift')
```

```python
# scrape cryptocurrency subreddit for comments mentioning 'loopring'

subreddit='cryptocurrency'
limit=300000
filters=['created_utc', 'body', 'id']
loopring_cc_comments = api.search_comments(q='loopring', subreddit=subreddit, filter=filters, limit=limit, before=before, after=after)
print(f'Retrieved {len(loopring_cc_comments)} comments from Pushshift')
```

```python
# scrape cryptocurrency subreddit for comments mentioning 'lrc'

subreddit="cryptocurrency"
limit=300000
filters=['created_utc', 'body', 'id']
lrc_cc_comments = api.search_comments(q='lrc', subreddit=subreddit, filter=filters, limit=limit, before=before, after=after)
print(f'Retrieved {len(lrc_cc_comments)} comments from Pushshift')
```

```python
# scrape superstonk subreddit for comments mentioning 'lrc'

subreddit="superstonk"
limit=300000
filters=['created_utc', 'body', 'id']
lrc_ss_comments = api.search_comments(q='lrc', subreddit=subreddit, filter=filters, limit=limit, before=before, after=after)
print(f'Retrieved {len(lrc_ss_comments)} comments from Pushshift')
```

```python
# comment data into DataFrames

loopring_ss_comments_df = pd.DataFrame(loopring_ss_comments)
lrc_ss_comments_df = pd.DataFrame(lrc_ss_comments)
loopring_cc_comments_df = pd.DataFrame(loopring_cc_comments)
lrc_cc_comments_df = pd.DataFrame(lrc_cc_comments)
```

```python
# comment DataFrames into csv

loopring_ss_comments_df.to_csv('./loopring_ss_comments.csv', header=True, index=False)
lrc_ss_comments_df.to_csv('./lrc_ss_comments.csv', header=True, index=False)
loopring_cc_comments_df.to_csv('./loopring_cc_comments.csv', header=True, index=False)
lrc_cc_comments_df.to_csv('./lrc_cc_comments.csv', header=True, index=False)
```
